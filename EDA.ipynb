```python

def columnas_con_un_valor(df):
    # Identificar las columnas con un solo valor único
    columnas_unicas = df.nunique()[df.nunique() == 1].index
    return columnas_unicas
```

```python
# LandCover
resultado1 = columnas_con_un_valor(Land_cover)

# Mostrar las columnas con un solo valor único
print(f"Columnas con un solo valor único: {resultado1}")
```
